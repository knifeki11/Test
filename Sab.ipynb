```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the provided CSV file
file_path = 'Updated_File (3).csv'  # Update this to match your uploaded file name
df = pd.read_csv(file_path)

# Display the first few rows to understand the data
display(df.head())

# Display the column names
display(df.columns)

# Display information about the DataFrame to understand the data types and missing values
display(df.info())

# Generate Plots

# Histogram for numerical data
df.hist(bins=30, figsize=(20, 15))
plt.show()

# Box Plot for numerical data
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()

# Scatter Plot between two columns (example)
plt.figure(figsize=(10, 6))
plt.scatter(df['AGE'], df['PH SALIVAIRE'])  # Replace with relevant column names
plt.xlabel('AGE')
plt.ylabel('PH SALIVAIRE')
plt.show()

# Correlation Heatmap (only numeric columns)
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()

# Bar Plot for a categorical column (example)
plt.figure(figsize=(12, 6))
sns.countplot(x='SEXE', data=df)  # Replace with a relevant categorical column
plt.xticks(rotation=90)
plt.show()



# Create a new column 'PH_Category' to categorize pH values
df['PH_Category'] = df['PH SALIVAIRE'].apply(lambda x: 'More than 7' if x > 7 else '7 or Less')

# Display the first few rows to see the new column
display(df.head())

# Generate a count plot for the 'PH_Category' column
plt.figure(figsize=(10, 6))
sns.countplot(x='PH_Category', data=df)
plt.title('Count of Cases with pH More Than 7 vs 7 or Less')
plt.xlabel('pH Category')
plt.ylabel('Count')
plt.show()

```
